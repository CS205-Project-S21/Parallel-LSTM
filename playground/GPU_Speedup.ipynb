```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
from IPython import display
```

```python
display.set_matplotlib_formats('svg')
```

```python
withoutGPU = np.array([3.45,2.34,1.27,1.43])
withGPU = np.array([1.12,1.10,0.12,0.13])
```

```python
name = ['Large Data \n Model 1', 'Large Data \n Model 2', 'Small Data \n Model 1', 'Small Data \n Model 2']
```

```python
x = np.arange(len(name))  # the label locations
width = 0.35  # the width of the bars


#fig = plt.figure(figsize=(8, 6)) 
#gs = gridspec.GridSpec(2, 1, height_ratios=[2, 1])
fig, (ax0, ax1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]}, figsize=[8,6], sharex=True)

#ax0 = plt.subplot(gs[0])
rects1 = ax0.bar(x - width/2, withoutGPU, width, label='Without GPU', color='royalblue')
rects2 = ax0.bar(x + width/2, withGPU, width, label='With GPU', color='darkorange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax0.set_ylabel('Time (s) / Epoch ', fontsize=15)
ax0.set_title('GPU v.s. No GPU Training Time',fontsize=15)

ax0.legend(fontsize=15)

#ax1 = plt.subplot(gs[1])
ax1.plot(withoutGPU/withGPU, 'ro-',)
ax1.set_ylabel('SpeedUp', fontsize=15)
ax1.set_xticks(x)
ax1.set_xticklabels(name, fontsize=15)

#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.savefig('../docs/GPUSpeedUp.png',dpi=400,transparent=True,bbox_inches='tight')
plt.show()
```
